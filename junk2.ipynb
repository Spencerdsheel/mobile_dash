```python
import os
import redis
import json
from celery import shared_task
import psycopg2
from datetime import datetime

# Redis config
redis_client = redis.Redis(host='localhost', port=6379, db=0)

# Postgres config
def get_pg_connection():
    return psycopg2.connect(
        dbname=os.getenv('DB_NAME'),
        user=os.getenv('DB_USER'),
        password=os.getenv('DB_PASSWORD'),
        host=os.getenv('DB_HOST'),
        port=os.getenv('DB_PORT')
    )

@shared_task
def update_redis_cache():
    try:
        # Get the last cached created_at value
        last_created_at = redis_client.get("created_at")
        if last_created_at:
            last_created_at = datetime.strptime(last_created_at.decode(), "%Y-%m-%d %H:%M:%S")
        else:
            # If not set, use a far-past date to get initial data
            last_created_at = datetime(2023, 11, 23, 0, 0, 0)

        print(f"Last cached at: {last_created_at}")

        conn = get_pg_connection()
        cursor = conn.cursor()

        # Fetch only new data
        cursor.execute("""
            SELECT id, booking_date, booking_from, booking_to, coach_type_name, fultterwave_charge, gsd_cov_fee, " \
            "gsd_tkt_rev,icrc_tkt_rev, insurance, medical, nrc_cov_fee, nrc_tkt_rev, no_of_passengers, pnr_number, seat_type, " \
            "stamp_duty, booking_status, total_cov_fee, total_fare, total_tkt_revenue, train_name, travel_date, type, user_name," \
            "user_type, created_at, route_name, poc_corporation_name FROM booking
            WHERE created_at > %s
            ORDER BY created_at ASC
        """, (last_created_at,))
        rows = cursor.fetchall()

        print(f"Fetched {len(rows)} new rows.")

        for row in rows:
            id_,booking_date, booking_from, booking_to, coach_type_name, fultterwave_charge, gsd_cov_fee, gsd_tkt_rev,icrc_tkt_rev, insurance, medical, nrc_cov_fee, nrc_tkt_rev, no_of_passengers, pnr_number, seat_type, stamp_duty, booking_status, total_cov_fee, total_fare, total_tkt_revenue, train_name, travel_date, type, user_name,user_type, created_at, route_name, poc_corporation_name = row


            # Check if already cached
            if redis_client.sismember("cache_index", id_):
                continue  # Skip duplicates

            # Save to Redis
            redis_key = f"data:{id_}"
            redis_value = {
                "id": id_,
                "booking_date": booking_date.strftime("%Y-%m-%d %H:%M:%S"), 
                "booking_from": booking_from, 
                "booking_to": booking_to, 
                "coach_type_name": coach_type_name, 
                "fultterwave_charge": fultterwave_charge, 
                "gsd_cov_fee": gsd_cov_fee, 
                "gsd_tkt_rev": gsd_tkt_rev,
                "icrc_tkt_rev": icrc_tkt_rev, 
                "insurance": insurance, 
                "medical": medical, 
                "nrc_cov_fee": nrc_cov_fee, 
                "nrc_tkt_rev": nrc_tkt_rev, 
                "no_of_passengers": no_of_passengers, 
                "pnr_number": pnr_number, 
                "seat_type": seat_type, 
                "stamp_duty": stamp_duty, 
                "booking_status": booking_status, 
                "total_cov_fee": total_cov_fee, 
                "total_fare": total_fare, 
                "total_tkt_revenue": total_tkt_revenue, 
                "train_name": train_name, 
                "travel_date": travel_date.strftime("%Y-%m-%d %H:%M:%S"), 
                "type": type, 
                "user_name": user_name,
                "user_type": user_type, 
                "created_at": created_at.strftime("%Y-%m-%d %H:%M:%S"), 
                "route_name": route_name, 
                "poc_corporation_name": poc_corporation_name,
            }
            redis_client.set(redis_key, json.dumps(redis_value))
            redis_client.sadd("cache_index", id_)

            # Update latest created_at if this is the newest so far
            redis_client.set("latest_created_at", created_at.strftime("%Y-%m-%d %H:%M:%S"))

        cursor.close()
        conn.close()

        return f"Cached {len(rows)} new records."

    except Exception as e:
        print("Error during caching:", e)
        return str(e)

```

```python
import json
from datetime import date

def generate_transaction_payload(ticket_id, passengers, transaction_date=None):
    # Constants
    BASE_PRICE_PER_PASSENGER = 3600
    MEDICAL_PER_PASSENGER = 30
    INSURANCE_PER_PASSENGER = 10
    STAMP_DUTY = 50
    CONVENIENCE_FEE_PERCENT = 0.065

    GSD_PERCENT = 0.185
    NRC_PERCENT = 0.795
    ICRC_PERCENT = 0.02
    CONVENIENCE_GSD_PERCENT = 0.60
    CONVENIENCE_NRC_PERCENT = 0.40

    # Defaults
    if transaction_date is None:
        transaction_date = str(date.today())

    # Base Calculations
    base_total = BASE_PRICE_PER_PASSENGER * passengers
    medical_total = MEDICAL_PER_PASSENGER * passengers
    insurance_total = INSURANCE_PER_PASSENGER * passengers
    convenience_fee_total = round(base_total * CONVENIENCE_FEE_PERCENT, 2)

    # Total amount
    total_amount = base_total + medical_total + insurance_total + STAMP_DUTY + convenience_fee_total

    # Split convenience fee
    conv_fee_gsd = round(convenience_fee_total * CONVENIENCE_GSD_PERCENT, 2)
    conv_fee_nrc = round(convenience_fee_total * CONVENIENCE_NRC_PERCENT, 2)

    # Party Distribution
    gsd_share = round((base_total + medical_total + insurance_total + STAMP_DUTY) * GSD_PERCENT + conv_fee_gsd, 2)
    nrc_share = round((base_total + medical_total + insurance_total + STAMP_DUTY) * NRC_PERCENT + conv_fee_nrc, 2)
    icrc_share = round((base_total + medical_total + insurance_total + STAMP_DUTY) * ICRC_PERCENT, 2)

    payload = {
        "date": transaction_date,
        "transactions": [
            {
                "ticket_id": ticket_id,
                "passengers": passengers,
                "base_price_per_passenger": BASE_PRICE_PER_PASSENGER,
                "convenience_fee": convenience_fee_total,
                "total_amount": total_amount,
                "split": {
                    "base_total": base_total,
                    "medical": medical_total,
                    "insurance": insurance_total,
                    "stamp_duty": STAMP_DUTY,
                    "convenience_fee": {
                        "total": convenience_fee_total,
                        "gsd_share": conv_fee_gsd,
                        "nrc_share": conv_fee_nrc
                    }
                },
                "parties_distribution": {
                    "gsd": gsd_share,
                    "nrc": nrc_share,
                    "icrc": icrc_share,
                    "medical": medical_total,
                    "insurance": insurance_total,
                    "stamp_duty": STAMP_DUTY,
                    "convenience_fee_total": convenience_fee_total
                }
            }
        ]
    }

    return payload

# 🔽 Example Usage
if __name__ == "__main__":
    result = generate_transaction_payload(ticket_id="TXN1002", passengers=2, transaction_date="2025-04-16")
    print(json.dumps(result, indent=2))
```

```python
import os
import math
from django.core.cache import cache
import psycopg2
import pandas as pd
import logging

# Logging setup
logging.basicConfig(level=logging.DEBUG)
logger = logging.getLogger(__name__)

CHUNK_SIZE = 10000  # You can experiment with this

def get_cached_dashboard_data():
    print("Checking dashboard data chunks in cache...")

    num_chunks = cache.get('dashboard_data_chunk_count')

    if num_chunks:
        print(f"Found {num_chunks} chunks in cache. Fetching...")
        all_data = []
        for i in range(num_chunks):
            chunk = cache.get(f'dashboard_data_chunk_{i}')
            if chunk:
                all_data.extend(chunk)
        return all_data

    print("Cache is empty, fetching data from DB...")

    try:
        connection = psycopg2.connect(
            dbname=os.getenv('DB_NAME'),
            user=os.getenv('DB_USER'),
            password=os.getenv('DB_PASSWORD'),
            host=os.getenv('DB_HOST'),
            port=os.getenv('DB_PORT')
        )

        cursor = connection.cursor()

        # Main query
        query = """SELECT 
            b.id, b.booking_date, b.booking_from, b.booking_id, b.booking_to,
            b.coach_type_name, b.fultterwave_charge, b.gsd_cov_fee, b.gsd_tkt_rev,
            b.icrc_tkt_rev, b.insurance, b.medical, b.nrc_cov_fee, b.nrc_tkt_rev,
            b.no_of_passengers, b.pnr_number, b.seat_type, b.stamp_duty,
            b.booking_status, b.total_cov_fee, b.total_fare, b.total_tkt_revenue,
            b.train_name, b.travel_date, b.type, b.user_name, b.user_type,
            b.created_at AS booking_created_at, b.updated_at AS booking_updated_at,
            b.route_name, b.poc_corporation_name,
            bd.id AS user_detail_id, bd.booking_id AS user_booking_id,
            bd.coach_name, bd.passenger_identification_number, bd.passenger_contact,
            bd.passenger_email, bd.passenger_name, bd.passenger_type, bd.seat_number,
            bd.seat_type AS user_seat_type, bd.created_at AS user_created_at,
            bd.updated_at AS user_updated_at
        FROM booking b
        LEFT JOIN booking_details bd ON bd.booking_id = b.booking_id
        ORDER BY b.booking_date;"""

        cursor.execute(query)
        data = cursor.fetchall()
        columns = [desc[0] for desc in cursor.description]

        df = pd.DataFrame(data, columns=columns)
        data_dict = df.to_dict(orient='records')

        # Chunk and store
        total = len(data_dict)
        num_chunks = math.ceil(total / CHUNK_SIZE)

        for i in range(num_chunks):
            chunk_data = data_dict[i * CHUNK_SIZE : (i + 1) * CHUNK_SIZE]
            cache.set(f'dashboard_data_chunk_{i}', chunk_data, timeout=None)

        cache.set('dashboard_data_chunk_count', num_chunks, timeout=None)

        logger.debug(f"{num_chunks} data chunks cached successfully.")
        cursor.close()
        connection.close()

        return data_dict

    except Exception as e:
        logger.error(f"DB fetch error: {e}")
        return "Failed to fetch data."
```

```python
import dash
from dash import dcc, html, Input, Output, State
import pandas as pd
from django.core.cache import cache

# --- Setup your Dash app as usual ---
app = dash.Dash(__name__, suppress_callback_exceptions=True)
app.layout = html.Div([
    html.Button("Refresh Dashboard", id="refresh-button"),
    html.Div(id="dashboard-content")
])

@app.callback(
    Output("dashboard-content", "children"),
    Input("refresh-button", "n_clicks")
)
def update_dashboard(n_clicks):
    # --- Step 1: Load Cached Chunks ---
    num_chunks = cache.get("dashboard_data_chunk_count")
    if not num_chunks:
        return html.Div("⚠️ No cached dashboard data available. Please run data fetch job.")

    all_records = []
    for i in range(num_chunks):
        chunk = cache.get(f"dashboard_data_chunk_{i}")
        if chunk:
            all_records.extend(chunk)

    if not all_records:
        return html.Div("⚠️ Dashboard data is empty.")

    dff = pd.DataFrame(all_records)

    # --- Step 2: Compute KPIs ---
    kpis = {
        "Tickets": dff['no_of_passengers'].sum(),
        "Transactions": len(dff),
        "Validated": dff[dff['booking_status'] == 'VALIDATED'].shape[0],
        "Online Sales": dff[dff['type'] == 'ONLINE']['total_fare'].sum(),
        "TOM Sales": dff[dff['type'] == 'OFFLINE']['total_fare'].sum(),
        "Total Sales": dff['total_fare'].sum(),
        "NRC Revenue": dff['nrc_tkt_rev'].sum()
    }

    kpi_cards = html.Div([
        html.Div(f"{key}: {int(val):,}", style={'padding': '10px', 'fontWeight': 'bold'})
        for key, val in kpis.items()
    ])

    # --- Step 3: Example Graph (Daily Fare) ---
    if 'booking_date' in dff.columns:
        dff['booking_date'] = pd.to_datetime(dff['booking_date'])
        daily_df = dff.groupby('booking_date')['total_fare'].sum().reset_index()

        daily_fig = {
            'data': [{
                'x': daily_df['booking_date'],
                'y': daily_df['total_fare'],
                'type': 'line',
                'name': 'Daily Fare'
            }],
            'layout': {
                'title': 'Daily Sales Over Time',
                'xaxis': {'title': 'Date'},
                'yaxis': {'title': 'Total Fare'}
            }
        }
    else:
        daily_fig = {}

    return html.Div([
        html.H3("📊 KPI Summary"),
        kpi_cards,
        html.Hr(),
        html.H3("📈 Daily Sales Graph"),
        dcc.Graph(figure=daily_fig)
    ])
```
